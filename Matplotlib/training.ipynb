### 完善折线图

```python
# 画出西安市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度-18度
import random

plt.rcParams['font.sans-serif']=['SimHei']

# 1、准备数据 x y
x = range(60)
y_xian= [random.uniform(15, 18) for i in x]
y_ankang = [random.uniform(18, 22) for i in x]

# 2、创建画布
plt.figure(figsize=(20,8), dpi=100)

# 3、绘制图像
plt.plot(x, y_xian, color='r', label='安康')
plt.plot(x, y_ankang, color='b', linestyle='--', label='西安')

# 修改x、y刻度
#准备x刻度说明
x_label = ['11点{}分'.format(i) for i in x]
plt.xticks(x[::5], x_label[5::5])
plt.yticks(range(0,41, 5))

# 添加网格
plt.grid(linestyle='--', alpha=0.5)

# 添加描述信息
plt.xlabel('时间变化')
plt.ylabel('温度变化')
plt.title('安康、西安11点到12点每分钟温度变化状况')

plt.legend(loc='best')

# 4、显示图像
plt.show()
```

### 多个坐标系画图

```python
# 画出西安市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度-18度
import random

plt.rcParams['font.sans-serif']=['SimHei']

# 1、准备数据 x y
x = range(60)
y_xian= [random.uniform(15, 18) for i in x]
y_ankang = [random.uniform(18, 22) for i in x]

# 2、创建画布
# plt.figure(figsize=(20,8), dpi=100)
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)

# 3、绘制图像
axes[0].plot(x, y_xian, color='r', label='安康')
axes[1].plot(x, y_ankang, color='b', linestyle='--', label='西安')

# 设置图例
axes[0].legend(loc='best')
axes[1].legend(loc='best')

# 修改x、y刻度
#准备x刻度说明
x_label = ['11点{}分'.format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label[::5])
axes[0].set_yticks(range(0,41, 5))
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(0,41, 5))

# 添加网格
axes[0].grid(linestyle='--', alpha=0.5)
axes[1].grid(linestyle='--', alpha=0.5)

# 添加描述信息
axes[0].set_xlabel('时间变化')
axes[0].set_ylabel('温度变化')
axes[0].set_title('安康11点到12点每分钟温度变化状况')
axes[1].set_xlabel('时间变化')
axes[1].set_ylabel('温度变化')
axes[1].set_title('西安11点到12点每分钟温度变化状况')

# 4、显示图像axes[0]
plt.show()
```

### 画数学函数

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

# 1、准备数据
x = np.linspace(-10, 10, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 2、创建画布
plt.figure(figsize=(20,8), dpi=120)

# 3、绘制函数图像
plt.plot(x, y_sin, color='r',label='sin')
plt.plot(x, y_cos, color='b',label='cos')
plt.legend(loc='best')

# 添加网格
plt.grid(linestyle='--', alpha=0.5)

plt.show()
```
