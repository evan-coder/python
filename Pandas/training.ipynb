```python
# 1、准备数据
data = pd.Series([165,174,160,180,159,163,192,184], index=['No1:165', 'No2:174','No3:160', 'No4:180', 'No5:159', 'No6:163', 'No7:192', 'No8:184'])
```

```python
data
```

```python
# 2、分组
# 自动分组
sr = pd.qcut(data, 3)
```

```python
type(sr)
```

```python
sr
```

```python
pd.get_dummies(sr, prefix='height')
```

```python
sr.value_counts()
```

```python
bins = [150, 165, 180, 195]
sr1 = pd.cut(data, bins)
```

```python
sr1
```

```python
sr.value_counts()
```

```python
pd.get_dummies(sr1, prefix='身高')
```

```python
# 1、读取数据
stock = pd.read_csv('stock_day.csv')
p_change = stock['p_change']
```

```python
p_change.head()
```

```python
# 2、分组
sr = pd.qcut(p_change, 10)
```

```python
sr
```

```python
# 3、离散化
pd.get_dummies(sr, prefix='涨跌幅')
```

```python
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
sr1 = pd.cut(p_change, bins=bins)
```

```python
sr1
```

```python
sr1.value_counts()
```

```python
pd.get_dummies(sr1, prefix='涨跌幅')
```
